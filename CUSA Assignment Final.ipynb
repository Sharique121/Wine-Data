```python
data2=data1.astype(object)
data2=pd.get_dummies(data2)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
sil=[]
sumsquare=[]
for i in range(3,6):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(data1)
    sil.append(silhouette_score(data1,kmeans.labels_))
    sumsquare.append(kmeans.inertia_)
```

```python
sil
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(data1)
```
